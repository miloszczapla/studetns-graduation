```python
import pandas as pd

# Specify the path to the file in the input directory
file_path = '/kaggle/input/student-graduation-dataset/graduation_dataset.csv'

# Read the file into a DataFrame
df = pd.read_csv(file_path)

# Now, 'df' contains the data from the 'graduation_dataset.csv' file
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
df.head(3)
```

```python
# Checking data types of each columns.
df.dtypes
```

```python
#### We should change Target into numbers.
df.Target.unique()
```

```python
# We are encoded Dropout , Graduate ,  Enrolled into numbers like 0,1,2
target_variable = df['Target']
# Initialize the LabelEncoder
label_encoder = LabelEncoder()
# Fit the LabelEncoder to the 'Target' column and transform it
df['Target'] = label_encoder.fit_transform(target_variable)
```

```python
df.head(3)
```

```python
# Checking whether there are missing values or not.
df.isnull().sum()
```

```python
# Assigning features and target variable.
features = df.iloc[:, :-1].to_numpy()
target = df['Target'].to_numpy().reshape(-1,1)
```

```python
# Checking the shape of the features.
features.shape
```

```python
# Checking the shape of the target.
target.shape
```

```python
# We have 3 classes so the labels will have 3 values.
# first class (1,0,0) second class (0,1,0) third class (0,0,1)
encoder = OneHotEncoder()
targets = encoder.fit_transform(target).toarray()
```

```python
# Spliting data set into training and testing.
feature_train,feature_test,target_train,target_test = train_test_split(features,targets,test_size=0.2)
```

```python
# Creating ANN.
model = Sequential()
model.add(Dense(10,input_dim=34,activation='relu')) # Input layer 34 neu, hidden layer 10neu.
model.add(Dense(10,input_dim=10,activation='relu'))
model.add(Dense(10,input_dim=10,activation='relu'))
model.add(Dense(3,activation='softmax')) # Output layer with 3 neurons. Softmax as an activation function.
```

```python
# We can define the loss function MSE or negative log likelihood.
# Optimizer will find the right adjustements for weights : SGD , ADAM , ADAGRAD. 
optimizer = Adam(learning_rate=0.001)

model.compile(loss='categorical_crossentropy', 
              optimizer=optimizer, 
              metrics=['accuracy'])

model.fit(feature_train, target_train, epochs=1000, batch_size=20, verbose=2)
results = model.evaluate(feature_test, target_test, use_multiprocessing=True)

print(results)
```

**In summary, this output provides information about the training process, including the number of epochs, training and testing loss, and training and testing accuracy. It indicates that the model was trained for 1000 epochs, achieved a training accuracy of approximately 81.60%, but the testing accuracy on unseen data was approximately 73.45%. This suggests that the model may be overfitting the training data to some extent, as the training accuracy is significantly higher than the testing accuracy.**
